# Welcome to Pythonland, the RPG!

Run all the code below to play. The "playGame()" section starts the game. 

If you want to stop the game and start again, press the "Stop" button (square above).

```python
def getInput(message,validanswers):
    isValid=False
    while not isValid:
        choice=input(message)
        if choice in validanswers:
            isValid=True
        else:
            print("I'm sorry, I didn't understand. Please answer again...")
    return choice
```

```python
from random import shuffle
from random import randint

def playGame():   
    room = "Intro"
    playerCheat = False
    stillPlaying = True
    stillAlive = True
    playerLevel=1
    playerClass = "Nitwit"
    firstDoors = ["beastRoom","teleportationRoom","randomRoom"]
    shuffle(firstDoors)
    if playerCheat:
        print("The doors are: " + ", ".join(firstDoors))
    while stillPlaying:
        if room =="Intro":
            introduction = '''Hello and welcome to the Pythonland Adventure!
The kingdom of Pythonland needs your help! The evil Lord Evilton has stolen the King's gold, 
and we need your help to get it back!'''
            print(introduction)
            choice = getInput("Do you accept the challenge? (y/n) ",["y","n"])
            if choice == "y":
                room = "Entrance"
            else:
                print("you return to your home. You are not a hero.")
                stillPlaying = False
        elif room == "Entrance":
            print ("""A powerful knight guards the door.""")
            choice = getInput("What do you do?\n1. Stab the knight\n2. Fireball the knight\n",["1","2"])
            if choice == "1":
                playerClass = "fighter"
            else:
                playerClass = "wizard"
            print("Your class is: "+ playerClass)
            room = "Inside"
        elif room == "Inside":
            print("Three identical doors lie before you")
            choice = getInput("Which door do you go through?\n1 Left\n2 Middle\n3 Right\n",["1","2","3"])
            room = firstDoors[int(choice)-1]
        elif room == "beastRoom":
            print("A red dragon sits in front of you on a pile of gold. A door at the far end of the room is locked shut ")
            if playerClass == "wizard":
                choice = getInput("What do you do?\n1. Run away with your feeble legs\n2.Teleport away\n3.Zap the dragon with your powerfullest spell",["1","2","3"])
            if playerClass == "fighter":
                choice = getInput("What do you do?\n1.Stab the dragon with your feeble copper shortsword\n2.Sprint away\n3.Leap over the giant dragon and stab it from behind.",["1","2","3"])
            if choice == "1":
                print("Youre eaten by the dragon. What did you expect?")
                stillPlaying = False
            if choice == "2":
                print("You escape by the skin of your teeth. You are still alive but you have gained no progress...")
                room = "Inside"
            if choice == "3":
                if playerLevel == 1:
                    print("The dragon is to powerful, and burns you up")
                    stillPlaying = False
                else:
                    print("You slay the dragon with ease and use the key you found with the kobolds to continue.")
                    room="Lair"
        elif room == "teleportationRoom":
            print("The room you enter teleports you back to the doors! They spin and rotate leaving you feeling lost.")
            room = "Inside"
            shuffle(firstDoors)
        elif room == "randomRoom":
            print("Dead goblins litter the floor.As you wonder why they are dead, a group of kobolds quickly jumps out of the shadows")
            choice = getInput("what do you do?\n1.Run away from the nimble kobolds\n2.Kill them all!",["1","2"])
            if choice == "1":
                print("They were faster than you, because you ran away, they slew you before you could attack")
                stillPlaying=False
            else:
                print("The kobolds are no match for you.They fall to the ground as you gain experience.Then a key drops from one of there pockets.It could be usefull...")
                playerLevel=2
                room="Inside"
        elif room == "Lair":
            print("Lord Evilton stands in front of you in all of his glory.He says,\"I was created using python, you cannot defeat me!\"")
            eviltonHealth=3
            playerHealth=3
            usedPowerful=False
            while playerHealth>0 and eviltonHealth>0:
                if playerClass == "wizard":
                    choice = getInput("What do you do?\n1.Use your last spell.\n2.Stab him\n3.Taunt him",["1","2","3"])
                if playerClass == "fighter":
                    choice = getInput("What do you do?\n1.Through your only poisoned dagger\n2.Punch him, your sword was destroyed by the dragon\n3.Heal!",["1","2","3"])
                if choice == "1":
                    if usedPowerful==False:
                        print("Evilton cries out in pain! But this was your last use...")
                        eviltonHealth=eviltonHealth-2
                        usedPowerful=True
                    else:
                        print("I told you you had used it, while you were distracted, Evilton snuck in another stab")
                        playerHealth=playerHealth-1
                if choice == "2":
                    if(randint(0, 1)==0):
                        eviltonHealth=eviltonHealth-1
                        if eviltonHealth>0:
                            print("Evilton looks slightly damaged from this.")
                        else:
                            print("The attack caught him by surprise!")
                    else:
                        print("Youre attack fully misses.Unlucky :)")
                        
                if choice == "3":
                    if(randint(0, 1)==0):
                        print("You heal above what evilton can do, whilst he is distracted")
                        playerHealth=playerHealth+2
                    else:
                        print("You fail to do anything.This is a risky buisness")
                if eviltonHealth > 0:
                    playerHealth=playerHealth-1
                    print("Evilton stabs you! Your health is " + str(playerHealth))
#            else:
#                print("You have entered an impossible room. The Dungeon Master is displeased. You die for no reason.")
#                stillPlaying=False
            if playerHealth>0:
                print("Lord Evilton is slain and Pythonland is saved!!!")
                stillPlaying=False
            else:
                print("Lord Evilton has killed you and Pythonland is trapped under his grasp, until another adventurer atempts to defeat him")
                stillPlaying=False
            
    

```
